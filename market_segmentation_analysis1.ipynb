```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
# Assuming the dataset has features that are relevant for market segmentation analysis
# Select the appropriate columns for analysis
selected_features = ['Two Wheeler','Three Wheeler','Four Wheeler']
print(selected_features)

# Extract the selected features
X = df[selected_features]

# Standardize the data for better clustering performance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters using the elbow method
inertia = []
max_clusters = 10  # Maximum number of clusters to try
for n_clusters in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow method graph to find the optimal number of clusters
plt.figure(figsize=(8, 6))
plt.plot(range(1, max_clusters + 1), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xticks(range(1, max_clusters + 1))
plt.grid()
plt.show()

# Based on the elbow method plot, choose the optimal number of clusters
num_clusters = 3  # Adjust this based on the graph and business requirements

# Apply K-means algorithm with the chosen number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X_scaled)

# Add the cluster labels to the original DataFrame
df['Cluster'] = kmeans.labels_

# Perform market segmentation analysis
cluster_summary = df.groupby('Cluster')[selected_features].mean()
print("Cluster Summary:")
print(cluster_summary)
```
